```python
#STEP 2 / DATA CLEANING AND TRANSORMATION 
duplicates = data.duplicated().sum()
print(f"duplicate values:{duplicates}")
if duplicates > 0:
     data = data.drop_duplicates()
     print(f"drop duplicated data{duplicates}")
else:
     print(f"no any duplicated data:{duplicates}")
```

```python
#Outlier detection (box plot )
#Numerical column outlier detection 

numerical_column = ["experience_years", "skills_count", "certifications", "salary"]
fig, axes = plt.subplots(1, 4, figsize=(18, 5))
fig.suptitle("outlier detection box plot")
for ax, col in zip(axes,numerical_column):
     sns.boxplot(y=data[col],ax=ax)
     ax.set_title(col)
plt.savefig("../Figures/outlier_boxplot.png")
plt.show()
```

```python
# outlier treatment using IQR method 
Q1 = data["salary"].quantile(0.25) #lower
Q3 = data["salary"].quantile(0.75) #upper
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outlier_counts = data[(data["salary"]<lower) | (data["salary"]>upper)]
print(f"salary outlier detected: {outlier_counts}")
print(f"lower bound: {lower:.0f}| upper bound: {upper:.0f}")

# Cap outliers at boundary values
data["salary"] = data["salary"].clip(lower=lower, upper=upper)
print("Outliers capped using IQR method")
```

### IQR = 0 indicates salary is uniformly distributed.This is characteristic of a synthetic dataset no outlier treatment needed because Q1 and Q3 are the exact same value (94,291)meaning salaries are spread so evenly and uniformly across the full range that the middle 50% has no variation at all.

```python
#Encode Categorical Variables 
# Encode categorical columns for analysis
from sklearn.preprocessing import LabelEncoder

# Education has natural order so we map it manually
education_order = {"High School": 0, "Bachelor": 1, "Master": 2, "PhD": 3}
data["education_encoded"] = data["education_level"].map(education_order)

# Remote work — label encode
le = LabelEncoder()
data["remote_work_encoded"] = le.fit_transform(data["remote_work"])

print("Encoding done")
print(data[["education_level", "education_encoded", "remote_work", "remote_work_encoded"]].head())
```
